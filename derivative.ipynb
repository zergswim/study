```python
import matplotlib.pyplot as plt
import numpy
import math

def drawGraphDe(labelParam, x, y):
    dx = x[1]-x[0]
    dydx = numpy.gradient(y, dx)
    plt.plot(x, y, label=labelParam)
    plt.plot(x, dydx, label=labelParam+' derivative')
    plt.legend()
    plt.show()

xs = numpy.linspace(0,10,1000)

ys = [x**2 + 1 for x in xs]
drawGraphDe('x**2+1', xs, ys)

ys = [math.sin(x) for x in xs]
drawGraphDe('sin(x)', xs, ys)

ys = [math.exp(x) for x in xs]
drawGraphDe('e**x', xs, ys)
```
